```r
library(ggplot2)
```

```r
projs= read.csv("combined_projects_commit_time_stat.csv")
summary(projs)
```

```r
ggplot(projs, aes(fill=group,x=as.factor(group),y=Mean))+
geom_boxplot(position = "dodge", notch="true") +
scale_y_continuous(trans='log2')+
geom_point()+
xlab("") + 
ylab (expression(paste("Average number of days"))) +
scale_fill_discrete(name = "Subject systems")+
theme(legend.position = "none")
ggsave("bg_projs_commit_time.pdf",width=6, height = 4, units = "in")
```
