```r
#load library
library(ggplot2)
```

```r
#load data
data=read.csv("RQ4_commit_time_computed.csv")
```

```r
summary(data)
```

```r
ggplot(data, aes(x = "", y = per_introDistance))+
geom_boxplot(notch = "true")+
xlab("")+
ylab("Introudction time (%)")+
geom_point()+
labs(xlab="", ylab="")
ggsave("RQ4_IntroTime_All.pdf",width=6, height = 4, units = "in")
```

```r
ggplot(data, aes(x = "", y = per_introDistance, fill=group))+
geom_boxplot(notch = "true")+
geom_point(width=0.2)
xlab("")+
ylab("Introudction time (%)")+
scale_fill_discrete(name = "Subject systems",labels=c("NoSQL","SQL"))+

theme(legend.position = "bottom")
ggsave("RQ4_IntroTime_SQLVsNoSQL.pdf",width=6, height = 4, units = "in")
```

<h3>Removal time box plot</h3>

```r
d=read.csv("RQ4_Removal_commit_time_computed.csv")
summary(d)
```

```r
#lets do the plot for remove time
ggplot(d, aes(x = "", y =  per_removeDistance, fill=group))+
geom_boxplot(notch = "FALSE")+
xlab("")+
ylab("Removal time (%)")+
scale_fill_discrete(name = "Subject systems",labels=c("NoSQL","SQL"))+

theme(legend.position = "bottom")
ggsave("RQ4_RemovalTime_SQLVsNoSQL.pdf",width=6, height = 4, units = "in")
```
