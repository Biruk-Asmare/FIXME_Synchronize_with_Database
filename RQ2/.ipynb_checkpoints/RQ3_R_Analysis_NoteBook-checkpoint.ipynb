<h1>Survivial Plots</h1>

```r
install.packages("ggpubr", dependencies = TRUE)
```

```r
#Load libraries
library(survival)
library(ggplot2)
library(ggpubr)
library("survminer")
```

```r
#Load data
data=read.csv('RQ3_SQL_Processed.csv')
nsdata=read.csv('RQ3_NOSQL_Processed.csv')
combined_sql=read.csv("RQ3_SQL_Combined_processed.csv")
combined_nsql=read.csv("RQ3_NOSQL_Combined_processed.csv")
```

<h2>KM for SQL subject systems</h2>

```r
ser=Surv(data$T, data$status)
kmsurv <-survfit(ser~1)
```

```r
str(kmsurv)
```

```r
#now using ggplot
plt <- ggsurvplot(kmsurv, data = data, pval = TRUE,size = 1,                 # change line size
surv.median.line = "hv",
  conf.int = TRUE, # Add confidence interval
legend="none",
  ggtheme = theme_gray(), 
        xlab="Number of commit units (1 unit=500 commits)"
          )
plt
ggsave("RQ2_Kaplan_Meir_SQL.pdf",width=6, height = 4, units = "in")
```

<h2>KM for NoSQL subject systems</h2>

```r
ser=Surv(nsdata$T, nsdata$status)
kmsurv <-survfit(ser~1)
```

```r
#now using ggplot
plt <- ggsurvplot(kmsurv, data = nsdata, pval = TRUE,size = 1,                 # change line size
surv.median.line = "hv",
  conf.int = TRUE, # Add confidence interval
legend="none",
 ggtheme = theme_gray(), 
        xlab="Number of commit units (1 unit=500 commits)"
          )
plt
ggsave("RQ2_Kaplan_Meir_NoSQL.pdf",width=6, height = 4, units = "in")
```

<h2>Combined survivial analysis plots SQL DAC vs NonDAC </h2>

```r
ser<-Surv(combined_sql$T, combined_sql$status)
kmsurv <-survfit(Surv(combined_sql$T, combined_sql$status)~combined_sql$type)   
summary(kmsurv)
```

```r
survdiff(formula = ser~combined_sql$type)
```

```r
#now using ggplot
plt <- ggsurvplot(kmsurv, data = combined_sql, 
pval = TRUE,size = 1,                 # change line size
pval.coord = c(30, 0.25),
surv.median.line = "hv",
conf.int = TRUE, # Add confidence interval
legend.title="",
legend.labs=c("Data-access","Non-data-access"),

  ggtheme = theme_gray(), 
        xlab="Number of commit units (1 unit=500 commits)"
          )
plt
ggsave("RQ2_Kaplan_Meir_SQL_DAC_NDAC.pdf.pdf",width=6, height = 4, units = "in")
```

<h2>Combined survivial analysis plots NoSQL DAC vs NonDAC </h2>

```r
ser<-Surv(combined_nsql$T, combined_nsql$status)
kmsurv <-survfit(Surv(combined_nsql$T, combined_nsql$status)~combined_nsql$type)   
summary(kmsurv)
```

```r
#now using ggplot
plt <- ggsurvplot(kmsurv, data = combined_nsql, 
pval = TRUE,size = 1,                 # change line size
pval.coord = c(4, 0.5),
surv.median.line = "hv",
conf.int = TRUE, # Add confidence interval
legend.title="",
legend.labs=c("Data-access","Non-data-access"),

  ggtheme = theme_gray(), 
        xlab="Number of commit units (1 unit=500 commits)"
          )
plt
ggsave("RQ2_Kaplan_Meir_NoSQL_DAC_NDAC.pdf.pdf",width=6, height = 4, units = "in")
```
