```r
library(ggplot2)
```

<h2>Distribution of commits in subject systems</h2>

```r
projs= read.csv("projects.csv")
summary(projs)
```

```r
ggplot(projs, aes(fill=group,x=as.factor(group),y=totalCommits))+
geom_boxplot(position = "dodge", notch="true") +
geom_point()+
xlab("") + 
ylab (expression(paste("Number of commits"))) +
scale_fill_discrete(name = "Subject systems")+
theme(legend.position = "none")
ggsave("RQ1_projs_boxplot.pdf",width=6, height = 4, units = "in")
```
